# Student Performance - Exploratory Data Analysis

Loads the raw dataset, checks data quality, handles missing values, and visualizes relationships between features and the pass/fail target.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
df = pd.read_csv("../data/students.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Missing values

```python
df.isnull().sum()
```

```python
# Median imputation for numeric columns with gaps
clean_df = df.copy()
for col in ["attendance_percentage", "sleep_hours"]:
    clean_df[col] = clean_df[col].fillna(clean_df[col].median())
clean_df = clean_df.drop_duplicates()
clean_df.isnull().sum()
```

## Class balance

```python
clean_df["passed"].value_counts(normalize=True).plot(kind="bar", title="Pass/Fail balance")
plt.xticks([0, 1], ["Fail", "Pass"], rotation=0)
plt.ylabel("Proportion")
plt.show()
```

## Feature distributions

```python
features = ["hours_studied", "attendance_percentage", "previous_grade", "sleep_hours", "assignments_completed"]
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
for ax, col in zip(axes.flat, features):
    sns.histplot(clean_df[col], kde=True, ax=ax)
    ax.set_title(col)
fig.delaxes(axes.flat[-1])
plt.tight_layout()
plt.show()
```

## Feature vs. target relationships

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
for ax, col in zip(axes.flat, features):
    sns.boxplot(data=clean_df, x="passed", y=col, ax=ax)
    ax.set_xticklabels(["Fail", "Pass"])
    ax.set_title(f"{col} by outcome")
fig.delaxes(axes.flat[-1])
plt.tight_layout()
plt.show()
```

## Correlation heatmap

```python
plt.figure(figsize=(7, 5))
sns.heatmap(clean_df[features + ["passed"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature correlation matrix")
plt.show()
```

```python
clean_df.to_csv("../data/students_clean.csv", index=False)
print("Saved cleaned dataset")
```
